### Creating DataFrame from CSV file

```python
d_parser = lambda x : pd.datetime.strptime(x, '%d-%m-%Y')
SpaceXdf = pd.read_csv('Spacex.csv',parse_dates=['Date'], date_parser=d_parser)
SpaceXdf['Date'] = SpaceXdf['Date'].dt.date
SpaceXdf.head()
```

### Renaming the Columns of the Pandas DataFrame

```python
SpaceXdf.rename({'Time (UTC)':'Time_UTC','PAYLOAD_MASS__KG_':'Payload_Mass_in_Kg', 'Landing _Outcome':'Landing_Outcome'}, axis=1, inplace=True)
SpaceXdf.head()
```

### Creating DataBase from Pandas DataFrame

```python
SpaceXdf.to_sql('SpaceXtable', conn, if_exists='replace', index=False)
```

```python
sql = ''' SELECT * FROM SpaceXtable'''
pd.read_sql(sql, conn)
```

## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
task_1 = '''
        SELECT DISTINCT Launch_Site 
        FROM SpaceXtable
        '''
pd.read_sql(task_1, conn)
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
task_2 = '''
        SELECT *
        FROM SpaceXtable
        WHERE Launch_Site LIKE 'CCA%'
        LIMIT 5
        '''
pd.read_sql(task_2, conn)
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
task_3 = '''
        SELECT SUM(Payload_Mass_in_Kg) AS Total_Payload_Mass_in_Kg
        FROM SpaceXtable
        WHERE Customer LIKE 'NASA (CRS)'
        '''
pd.read_sql(task_3, conn)
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
task_4 = '''
        SELECT AVG(Payload_Mass_in_Kg) AS Average_Payload_Mass_in_Kg
        FROM SpaceXtable
        WHERE Booster_Version = 'F9 v1.1'
        '''
pd.read_sql(task_4, conn)
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
task_5 = '''
        SELECT MIN(Date) AS First_Successful_Landing_Date
        FROM SpaceXtable
        WHERE Landing_Outcome LIKE 'Success (ground pad)'
        '''
pd.read_sql(task_5, conn)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
task_6 = '''
        SELECT Booster_Version
        FROM SpaceXtable
        WHERE Landing_Outcome = 'Success (drone ship)'
            AND Payload_Mass_in_Kg > 4000 
            AND Payload_Mass_in_Kg < 6000
        '''
pd.read_sql(task_6, conn)
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
task_7a = '''
        SELECT COUNT(Mission_Outcome) AS No_of_Successful_Outcome
        FROM SpaceXtable
        WHERE Mission_Outcome LIKE 'Success%'
        '''

task_7b = '''
        SELECT COUNT(Mission_Outcome) AS No_of_Failure_Outcome
        FROM SpaceXtable
        WHERE Mission_Outcome LIKE 'Failure%'
        '''
print('The total number of successful mission outcome is:')
display(pd.read_sql(task_7a,conn))
print()
print('The total number of failed mission outcome is:')
display(pd.read_sql(task_7b,conn))
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
task_8 = '''
        SELECT Booster_Version, Payload_Mass_in_Kg
        FROM SpaceXtable
        WHERE Payload_Mass_in_Kg = (
                                SELECT MAX(Payload_Mass_in_Kg)
                                FROM SpaceXtable
                                )
        ORDER BY Booster_Version
        '''
pd.read_sql(task_8,conn)
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
task_9 = '''
        SELECT Booster_Version, Launch_Site, Landing_Outcome
        FROM SpaceXtable
        WHERE Landing_Outcome LIKE 'Failure (drone ship)'
            AND Date BETWEEN '2015-01-01' AND '2015-12-31'
        '''
pd.read_sql(task_9,conn)
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
task_10 = '''
        SELECT Landing_Outcome, COUNT(Landing_Outcome)
        FROM SpaceXtable
        WHERE DATE BETWEEN '2010-06-04' AND '2017-03-20'
        GROUP BY Landing_Outcome
        ORDER BY COUNT(Landing_Outcome) DESC
        '''
pd.read_sql(task_10,conn)
```
